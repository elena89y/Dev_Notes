## 결측값 처리 (Handling Missing Values)

먼저 데이터에 포함된 결측값의 현황을 확인한다.  
`hotel_df`에는 다음과 같은 결측값이 존재한다.

- **children**: 4개  
- **country**: 488개  
- **agent**: 16,340개  
- **company**: 112,593개  

---

### 1. `children` 결측값 처리

`children` 컬럼의 결측값은 4개로 매우 적어  
전체 데이터에 미치는 영향이 크지 않다고 판단하였다.  

해당 결측값은 **어린이를 동반하지 않은 예약**으로 가정하고  
`0`으로 대체하여 분석을 진행한다.

---

### 2. `country` 결측값 처리

`country` 컬럼에는 488개의 결측값이 존재한다.  
국가 정보는 그 자체로 하나의 범주로 의미를 가질 수 있으므로,  

결측값을 제거하지 않고  
**`Unknown`이라는 별도의 카테고리로 대체**하여 처리한다.

---

### 3. `company` 컬럼 제거

`company` 컬럼은 전체 119,390건 중  
112,593건이 결측값으로, 결측 비율이 매우 높다.  

대부분 **개인 예약**으로 판단되며,  
분석에 유의미한 정보를 제공하기 어렵다고 판단하여  
해당 컬럼은 분석 대상에서 제외한다.

---

### 4. `agent` 결측값 처리

`agent` 컬럼의 결측값은  
**여행사를 통하지 않은 예약**으로 해석할 수 있다.  

여행사 예약과 비여행사 예약 간의 비교가 가능하도록,  
해당 결측값을 `0`으로 대체하여  
**범주형 변수로 활용**한다.

추후 이를 통해  
여행사 예약 여부에 따른 **취소율 차이**를 분석할 수 있다.

---

이와 같은 결측값 처리 전략을 통해  
데이터의 손실을 최소화하면서도  
분석에 필요한 정보는 최대한 유지하고자 한다.


```python
# children과 agent 결측치 0으로 대체
hotel_df['children'] = hotel_df['children'].fillna(0)
hotel_df['agent'] = hotel_df['agent'].fillna(0)

# country 결측치를 Unknown으로 대체
hotel_df['country'] = hotel_df['country'].fillna("Unknown")

# company 열 삭제
hotel_df = hotel_df.drop(columns='company')
```

**agent**의 경우 **agent ID** 이므로, <br>

```python
hotel_df['agent_group'] = hotel_df['agent'].apply(
    lambda x: 'No Agent' if x == 0 else 'With Agent'
)
```

## 데이터 정제

결측값 처리가 완료된 이후에는,  
데이터에 남아 있는 **논리적 오류, 이상치, 의미상 중복 및 누수 변수**를 추가로 정제한다.  
이 과정은 취소율 분석의 신뢰도를 높이기 위해 필수적인 단계이다.

---

### 1. 논리적으로 성립하지 않는 값 제거

예약이 존재함에도 불구하고 실제 상황으로 보기 어려운 경우를 먼저 제거한다.

- **숙박 인원이 전혀 없는 경우**
  - `adults == 0` AND `children == 0` AND `babies == 0`

- **숙박 일수가 0인 경우**
  - `stays_in_week_nights + stays_in_weekend_nights == 0`

위 조건에 해당하는 데이터는 실제 예약 데이터로 보기 어렵다고 판단하여  
분석 대상에서 제외한다.

---

### 2. 이상치 처리

이상치는 특히 **취소율 분석에 큰 영향을 미칠 수 있으므로**  
사전에 확인하고 적절히 처리한다.

다음 변수들을 중심으로 이상치를 점검한다.

- **`lead_time`**  
  - 예약 시점부터 도착까지의 기간이 지나치게 큰 경우  
  - → 제거하거나 구간화(bin) 처리

- **`booking_changes`**  
  - 예약 변경 횟수가 비정상적으로 많은 경우  
  - → 상한 처리(capping) 또는 제거

- **`days_in_waiting_list`**  
  - 대기 기간이 극단적으로 긴 경우  
  - → 제거 또는 이진화/구간화 처리

이러한 처리를 통해 소수의 극단값이 분석 결과를 왜곡하는 것을 방지한다.

---

### 3. 의미상 중복 및 데이터 누수 변수 제거

다음 변수들은 **예약 취소 결과 이후에 결정되거나 결과 자체를 포함하는 변수**로,  
취소 원인 분석에 사용할 경우 데이터 누수를 유발할 수 있다.

- **`assigned_room_type`**
  - 체크인 직전 또는 이후에 결정되는 운영 결과
  - 데이터 누수

- **`reservation_status`**
  - 예약의 최종 결과 자체
  - 정답지에 해당

- **`reservation_status_date`**
  - 최종 상태가 기록된 사후 날짜
  - 미래 정보

위 변수들은 분석 및 모델링 대상에서 제외한다.

---

### 4. 범주형 변수 카테고리 정리 (재분류)

일부 범주형 변수는 의미가 중복되거나,  
관측 수가 매우 적어 분석 시 노이즈를 유발할 수 있다.

- **`meal`**
  - `Undefined`는 실제로 `SC`(식사 미포함)와 동일한 의미이므로 통합

- **`market_segment`, `distribution_channel`**
  - 희귀 카테고리가 다수 존재하여 취소율 계산 시 불안정
  - 제거할 경우 실제 예약 유형 정보를 손실할 수 있으므로  
    → 의미가 유사한 값은 통합하여 **정보는 유지하고 노이즈는 감소**

해당 변수들은 **제거가 아닌 재분류/통합** 대상으로 처리한다.


**!! 이러한 데이터 정제 과정을 통해  
분석 결과의 신뢰도와 해석 가능성을 높이고,  
예약 취소와 실제로 연관된 요인에 집중할 수 있도록 하였다.


```python
# 숙박 인원 0인 데이터 확인

invalid_guests = hotel_df[
    (hotel_df['adults'] == 0) &
    (hotel_df['children'] == 0) &
    (hotel_df['babies'] == 0)]

len(invalid_guests)
```

```python
# 숙박 일수 0인 데이터 확인

invalid_stays = hotel_df[
    (hotel_df['stays_in_week_nights'] + hotel_df['stays_in_weekend_nights']) == 0]

len(invalid_stays)
```

```python
# 위 두 조건 제거

hotel_df = hotel_df[
    ~(((hotel_df['adults'] == 0) &
        (hotel_df['children'] == 0) &
        (hotel_df['babies'] == 0)) |
        ((hotel_df['stays_in_week_nights'] +
        hotel_df['stays_in_weekend_nights']) == 0))]
```

```python
# 의미상 중복이 있거나 데이터 누수가 있을 법한 변수 제거

hotel_df = hotel_df.drop(columns=['reservation_status', 'reservation_status_date'])
```

## 1차 분리 분석: 누가 예약을 취소하는가?

취소율을 기준으로 고객을 1차적으로 분리하여 살펴본 결과,  
호텔 유형, 여행사 이용 여부, 예약 유통 채널에 따라  
취소율에 뚜렷한 차이가 나타났다.

---

### 1. 호텔 유형별 취소율

- **City Hotel**: 약 **41.9%**
- **Resort Hotel**: 약 **28.0%**

City Hotel의 취소율이 Resort Hotel보다  
상대적으로 높게 나타났다.  
이는 도심 호텔 예약이 출장, 단기 일정 등  
변동성이 높은 목적과 더 많이 연관되어 있기 때문으로 추정된다.

---

### 2. 여행사 이용 여부에 따른 취소율

- **여행사 미경유(No Agent)**: 약 **24.9%**
- **여행사 경유(With Agent)**: 약 **39.2%**

여행사를 통해 예약한 경우의 취소율이  
직접 예약한 경우보다 유의미하게 높게 나타났다.  
이는 여행사 예약이 상대적으로  
취소 정책이 유연하거나 단체·대기성 예약이 많기 때문일 가능성을 시사한다.

---

### 3. 예약 유통 채널별 취소율

- **Undefined**: **80.0%**
- **TA/TO**: 약 **41.2%**
- **Corporate**: 약 **22.2%**
- **GDS**: 약 **19.5%**
- **Direct**: 약 **17.6%**

유통 채널에 따라 취소율 차이가 매우 크게 나타났으며,  
특히 여행사 및 투어 오퍼레이터(TA/TO)를 통한 예약의 취소율이 높은 반면,  
직접 예약(Direct)과 기업 예약(Corporate)은 상대적으로 안정적인 모습을 보였다.

---

이러한 결과를 통해,  
**City Hotel**, **여행사 경유 예약**, **TA/TO 채널**이  
취소율 관리에 있어 우선적으로 주목해야 할 대상임을 확인할 수 있다.

## 2차 조건 분석: 어떤 조건에서 예약이 취소되는가?

1차 분석을 통해 취소율이 높은 집단을 확인한 이후,  
이번에는 예약의 조건과 상태에 따라  
취소율이 어떻게 달라지는지 살펴본다.

---

### 1. 예약 리드타임(lead_time)에 따른 취소율

| 예약 시점 ~ 도착 시점 | 취소율 |
|----------------------|--------|
| 1개월 미만 (<1M) | 약 **18.8%** |
| 1–3개월 | 약 **37.8%** |
| 3–6개월 | 약 **44.8%** |
| 6–12개월 | 약 **55.5%** |
| 1년 초과 (>1Y) | 약 **67.7%** |

예약 시점과 도착 시점 간의 간격이 길어질수록  
취소율이 **단조 증가하는 패턴**이 뚜렷하게 나타났다.

이는 장기 예약일수록  
일정 변경 가능성이 높고 불확실성이 커지기 때문으로 해석할 수 있다.

---

### 2. 예약 변경 횟수(booking_changes)에 따른 취소율

예약 변경 횟수에 따른 취소율을 살펴보면,  
단순히 변경 횟수가 많을수록 취소율이 증가하는 **단조 관계는 나타나지 않았다.**

- **변경이 전혀 없는 예약(0회)**의 취소율이 오히려 가장 높게 나타났으며  
- 일정 수준의 변경이 발생한 예약은  
  취소율이 상대적으로 낮은 경향을 보였다.

이는 예약을 변경하는 고객이  
예약을 유지하려는 의도가 상대적으로 강할 수 있음을 시사한다.

---

### 3. 대기 경험 여부(has_waiting)에 따른 취소율

- **대기 경험 없음 (0)**: 약 **36.4%**
- **대기 경험 있음 (1)**: 약 **63.9%**

대기자 명단에 포함된 경험이 있는 예약의 경우,  
취소율이 **현저히 높게 나타났다.**

대기 상태 자체가 예약의 불확실성을 반영하며,  
이로 인해 취소 가능성이 크게 증가하는 것으로 해석할 수 있다.

---

### 요약

2차 조건 분석 결과,

취소율과 강한 연관성을 가지는 핵심 조건으로 확인되었다.

반면 예약 변경 횟수는  
단순한 횟수 증가보다는 고객의 예약 유지 의도를 함께 고려해야 할 변수로 판단된다.

## 3차 고객 성향 분석: 어떤 사람이 예약을 취소하는가?

앞선 분석이 예약 조건과 외부 요인에 초점을 맞췄다면,  
이번 단계에서는 고객의 **성향과 행동 특성**에 따라  
취소율이 어떻게 달라지는지 살펴본다.

---

### 1. 반복 방문 여부(is_repeated_guest)에 따른 취소율

- **기존 방문 이력이 없는 고객(0)**의 취소율이  
  반복 방문 고객에 비해 더 높게 나타났다.
- **반복 방문 고객(1)**은 전반적으로  
  취소율이 낮아, 예약 신뢰도가 높은 집단으로 확인된다.

이는 호텔 이용 경험이 있는 고객일수록  
예약을 유지하려는 경향이 강하다는 점을 시사한다.

---

### 2. 특별 요청 건수(total_of_special_requests)에 따른 취소율

특별 요청 건수에 따른 취소율을 살펴본 결과,  
요청 건수가 **0건인 경우 취소율이 상대적으로 높게 나타났으며**,  
요청이 1건 이상 존재하는 경우 전반적으로 취소율이 낮아지는 경향을 보였다.

이는 특별 요청을 남긴 고객이  
예약에 대한 의사결정이 비교적 명확하고,  
투숙 의도가 더 확실한 고객일 가능성을 시사한다.

---

### 3. 식사 유형(meal)에 따른 취소율

| 식사 유형 | 취소율 |
|----------|--------|
| **HB** (Half Board) | 약 **34.6%** |
| **SC** (식사 미포함) | 약 **36.3%** |
| **BB** (조식 포함) | 약 **37.6%** |
| **FB** (Full Board) | 약 **59.9%** |

식사 유형에 따라 취소율 차이가 나타났으며,  
특히 **FB(Full Board)** 예약의 취소율이  
다른 식사 유형에 비해 현저히 높게 나타났다.

이는 풀보드 예약이  
장기 일정이나 패키지성 예약과 연관되어  
일정 변경이나 취소 가능성이 상대적으로 높기 때문으로 해석할 수 있다.

---
